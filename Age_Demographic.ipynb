```python
import csv
import spacy
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
import pandas as pd
import csv 
```

```python
file = open('/kaggle/input/harry-potter-dataset/Harry Potter 1.csv','r')
HP = csv.reader(file, delimiter=';')
```

```python
characters = ['Harry','Snape','Dumbledore','Voldemort', 'Hermione', 'Ron', 'Draco','McGonagall', 'Hagrid','Petunia', 'Neville', 'Dudley', 'Vernon', 'Quirrell']
```

```python
fields = ['dialogue','age']
rows = []
for line in HP:
  if(line[0]=='Harry' or line[0]=='Hermione' or line[0]=='Ron' or line[0]=='Draco' or line[0]=='Neville' or line[0]=='Dudley'):
    row = [line[1], 12]
    rows.append(row)
  elif(line[0]=='Snape'):
    row = [line[1], 32]
    rows.append(row)
  elif(line[0]=='Dumbledore'):
    row = (line[1], 80)
    rows.append(row)
  elif(line[0]=='Voldemort'):
    row = (line[1], 65)
    rows.append(row)
  elif(line[0]=='McGonagall'):
    row = (line[1], 56)
    rows.append(row)
  elif(line[0]=='Hagrid'):
    row = (line[1], 60)
    rows.append(row)
  elif(line[0]=='Petunia'):
    row = (line[1], 40)
    rows.append(row)
  elif(line[0]=='Vernon'):
    row = (line[1], 45)
    rows.append(row)
  elif(line[0]=='Vernon'):
    row = (line[1], 45)
    rows.append(row)
  elif(line[0]=='Quirrell'):
    row = (line[1], 20)
    rows.append(row)
  else:
    pass
```

```python
filename = "labeled_dialogues.csv"
with open(filename, 'w') as csvfile: 
    # creating a csv writer object 
    csvwriter = csv.writer(csvfile) 
        
    # writing the fields 
    csvwriter.writerow(fields) 
        
    # writing the data rows 
    csvwriter.writerows(rows)
```

```python
dataset = pd.read_csv("/kaggle/working/labeled_dialogues.csv")
train_data = dataset.sample(frac=0.8, random_state=42)
test_data = dataset.drop(train_data.index) 
```

```python
text_clf = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('clf', MultinomialNB())
])
```

```python
text_clf.fit(train_data['dialogue'], train_data['age'])
```

```python
predicted = text_clf.predict(test_data['dialogue'])
print(classification_report(test_data['age'], predicted))
```
